```python
import pandas as pd
import os
```

```python
# Compare runs

run_dict = {
    'Scenario A': r'C:\workspace\sc_18_bike_count_test',
    'Scenario B': r'C:\workspace\sc_18_bike_count_test',
    'Scenario C': r'C:\workspace\sc_18_bike_count_test'
}
```

```python
def load_data(fname):
    df = pd.DataFrame()
    for scen_name, scen_path in run_dict.items():
        _df = pd.read_csv(os.path.join(scen_path,r'outputs\compare\rgc',fname))
        _df['source'] = scen_name
        df = df.append(_df)
    
    return df
```

# Population

```python
pd.options.display.float_format = '{:0,.0f}'.format
df = load_data('1_population.csv')
df = df.pivot_table(index='hh_rgc', columns='source', aggfunc='sum', values='Persons')
```

# Employment

```python
df = load_data('2_employment.csv')
df['Workers'] = df['Paid Full-Time Worker']+df['Paid Part-Time Worker']
df = df.pivot_table(index='hh_rgc', columns='source', aggfunc='sum', values='Workers')
df
```

# Transit Ridership

```python
df = load_data('3_transit_trips.csv')
df = df.pivot_table(index='hh_rgc', columns='source', aggfunc='sum', values='trexpfac')
# df.plot(kind='barh', alpha=0.6)
df
```

## Transit Commute Mode Share

```python
pd.options.display.float_format = '{:0,.1%}'.format
df = load_data('4_transit_commute_share.csv')
df = df.pivot_table(index='hh_rgc', columns='source', aggfunc='sum', values='mode_share')
# df.plot(kind='barh', alpha=0.6)
df
```
